```python
# RESIZE
from PIL import Image
import os

# Define the paths to your input and output directories
input_dir = "/Users/sidhanthmantena/Desktop/sriram"
output_dir = "/Users/sidhanthmantena/Desktop/sriram1"

# Define the target size
target_size = (224, 224)  # Width x Height

# Loop through images in the input directory and resize them
for filename in os.listdir(input_dir):
    if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.JPG')):  # Check for supported image formats
        image_path = os.path.join(input_dir, filename)
        img = Image.open(image_path)
        img = img.resize(target_size, Image.LANCZOS)  # Use Image.ANTIALIAS as the filter
        output_path = os.path.join(output_dir, filename)
        img.save(output_path)

# Data processed
print("Data processed.")
```

```python
# RENAME
import os

# Set the path to the folder containing the files you want to rename
folder_path = '/Users/sidhanthmantena/Desktop/sriram1'

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Initialize a counter for renaming
counter = 1

# Iterate through the files and rename them
for filename in files:
    # Construct the new filename
    new_filename = f'image{counter}.jpg'  # You can change the file extension if needed
    
    # Build the full paths for the old and new filenames
    old_filepath = os.path.join(folder_path, filename)
    new_filepath = os.path.join(folder_path, new_filename)
    
    # Check if the destination file already exists
    while os.path.exists(new_filepath):
        # If it exists, generate a new unique name
        base_name, extension = os.path.splitext(new_filename)
        new_filename = f'{base_name}_{counter}{extension}'
        new_filepath = os.path.join(folder_path, new_filename)
    
    try:
        # Rename the file
        os.rename(old_filepath, new_filepath)
    except Exception as e:
        print(f"Error renaming {filename}: {str(e)}")
    
    # Increment the counter for the next file
    counter += 1

print("Files renamed successfully.")
```

```python
# SPLIT POSITIVES

import os
import random
import shutil

# Define the path to your dataset
base_dir = '/Users/sidhanthmantena/Desktop/pneumonia'

# Define the subfolder name for positive images
positive_folder_name = 'positives'

# Create folders for train, test, and validate sets
os.makedirs(os.path.join(base_dir, 'train', positive_folder_name), exist_ok=True)
os.makedirs(os.path.join(base_dir, 'test', positive_folder_name), exist_ok=True)
os.makedirs(os.path.join(base_dir, 'validate', positive_folder_name), exist_ok=True)

# Define the split ratios
train_ratio = 0.7
test_ratio = 0.15
validate_ratio = 0.15

# Set a seed for reproducibility
random.seed(42)

# Iterate through the images in the dataset
for image_file in os.listdir(base_dir):
    if image_file.endswith('.jpg') or image_file.endswith('.png'):
        # Determine which set this image belongs to
        rand = random.random()
        if rand < train_ratio:
            destination = 'train'
        elif rand < train_ratio + test_ratio:
            destination = 'test'
        else:
            destination = 'validate'

        # Create the class folder if it doesn't exist
        class_folder = os.path.join(base_dir, destination, positive_folder_name, image_file.split('_')[0])
        os.makedirs(class_folder, exist_ok=True)

        # Move the image to the appropriate folder
        source = os.path.join(base_dir, image_file)
        destination = os.path.join(class_folder, image_file)
        shutil.move(source, destination)

print("Dataset splitting completed.")
```
